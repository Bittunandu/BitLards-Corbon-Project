```python
intents = {
    "greeting": {
        "patterns": ["hi", "hello", "hey"],
        "response": "Hello! How can I help you?"
    },
    "nlp": {
        "patterns": ["what is nlp", "define nlp"],
        "response": "NLP stands for Natural Language Processing."
    },
    "bye": {
        "patterns": ["bye", "goodbye"],
        "response": "Goodbye! Have a nice day "
    }
}
```

```python
def preprocess(text):
  return text.lower().strip()
```

```python
def get_response(user_input):
  user_input = preprocess(user_input)

  for intent in intents:
    for pattern in intents[intent]["patterns"]:
      if pattern in user_input:
        return intents[intent]["response"]
    return "Sorry, I didn't understand that."
```

```python
print("Chatbot: Hi! Type 'bye' to exit.")

while True:
    user_input = input("You: ")

    response = get_response(user_input)
    print("Chatbot:", response)

    if user_input.lower() == "bye":
        break
```

```python
def tokenize(text):
    return text.lower().split()
tokenize("Hello NLP World")
```

```python
sentences = [
    "hi hello",
    "what is nlp",
    "bye goodbye"
]

vocab = set()
for s in sentences:
    vocab.update(tokenize(s))

vocab = list(vocab)
print(vocab)
```

```python
def bag_of_words(sentence, vocab):
    words = tokenize(sentence)
    return [1 if w in words else 0 for w in vocab]
bag_of_words("hi nlp", vocab)
```

```python
texts = [
    "hi hello",
    "what is nlp",
    "bye goodbye"
]
labels = ["greeting", "nlp", "bye"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)

model = LogisticRegression()
model.fit(X, labels)


def predict_intent(text):
    X_test = vectorizer.transform([text])
    return model.predict(X_test)[0]
predict_intent("hello")
```

```python
responses = {
    "greeting": "Hello! How can I help you?",
    "nlp": "NLP stands for Natural Language Processing.",
    "bye": "Goodbye! "
}

print("AI Chatbot Ready! Type 'bye' to exit.")

while True:
    user = input("You: ")
    intent = predict_intent(user)
    print("Bot:", responses[intent])

    if intent == "bye":
        break
```
